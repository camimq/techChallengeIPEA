```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pd.read_csv ('../baseANP.csv')
```

```python
df.info()
```

```python
display(df)
```

```python
df['data_coleta'] = pd.to_datetime(df['data_coleta'])
```

```python
df['ano'] = df['data_coleta'].dt.year
df['mês'] = df['data_coleta'].dt.month
```

```python
### 1. Produtos mais vendidos 2019 a 2023 (litros)

# Contando a quantidade de produtos vendidos
produtos_contagem = df['produto'].value_counts()

# Criando o gráfico
plt.figure(figsize=(10,6))
ax = sns.barplot(x=produtos_contagem.values, y=produtos_contagem.index, palette='viridis')
plt.title('Produtos mais vendidos 2019 a 2023 (litros)')
plt.xlabel('Quantidade Vendida (litros)')
plt.ylabel('Produto')

# Adicionando rótulos de dados
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

plt.show()
```

#### Produtos mais vendidos 2019 a 2023 (litros)
A Gasolina foi o produto mais vendido em termos de quantidade de transações, o que reflete sua alta demanda.

```python
### 2. Produtos mais vendidos 2019 a 2023 (R$)

# Agrupando e somando os dados
produtos_valor = df.groupby('produto')['valor_venda'].sum().sort_values(ascending=False)

# Criando o gráfico
plt.figure(figsize=(10,6))
ax = sns.barplot(x=produtos_valor.values, y=produtos_valor.index, palette='viridis')
plt.title('Produtos mais vendidos 2019 a 2023 (R$)')
plt.xlabel('Valor Vendido (R$)')
plt.ylabel('Produto')

# Adicionando rótulos de dados
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

plt.show()
```

Em termos de valor, a gasolina também lidera, seguida pelo diesel. Isso indica que, além de ser o mais consumido, a gasolina também gera maior receita.

```python
### 3. Produtos mais vendidos por Região 2019 a 2023 (R$)

# Agrupando e somando os dados
regiao_valor = df.groupby('regiao')['valor_venda'].sum().sort_values(ascending=False)

# Criando o gráfico
plt.figure(figsize=(10,6))
ax = sns.barplot(x=regiao_valor.values, y=regiao_valor.index, palette='viridis')
plt.title('Produtos mais vendidos por Região 2019 a 2023 (R$)')
plt.xlabel('Valor Vendido (R$)')
plt.ylabel('Região')

# Adicionando rótulos de dados
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

plt.show()
```

A região Sudeste apresenta o maior valor de vendas, o que pode ser atribuído à alta concentração populacional e ao número de veículos nesta região.

```python
### 4. Venda Histórica por Produto

# Agrupando os dados por ano e produto
venda_historica = df.pivot_table(values='valor_venda', index='produto', columns='ano', aggfunc='sum', fill_value=0)

# Criando o gráfico
plt.figure(figsize=(12,8))
ax = sns.heatmap(venda_historica, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Venda Histórica por Produto')
plt.xlabel('Ano')
plt.ylabel('Produto')

plt.show()
```

É possível observar tendências e variações anuais nas vendas de diferentes produtos, destacando períodos de alta e baixa. Sendo em primeiro lugar a Gasolina como mais vendida e em sexto lugar o Etanol como menos vendido olhando de forma consolidada entre 2019 á 2023.

```python
### 5. Top 10 Revendas 2019 a 2023 (R$)

# Agrupando e somando os dados
revendas_valor = df.groupby('revenda')['valor_venda'].sum().sort_values(ascending=False).head(10)

# Criando o gráfico
plt.figure(figsize=(10,6))
ax = sns.barplot(x=revendas_valor.values, y=revendas_valor.index, palette='viridis')
plt.title('Top 10 Revendas 2019 a 2023 (R$)')
plt.xlabel('Valor Vendido (R$)')
plt.ylabel('Revenda')

# Adicionando rótulos de dados
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

plt.show()
```

Entre os revendedores em destaque esses são os top 10 que mais tiveram foco em estratégias de mercado, negócio e análise de desempenho durante os anos para liderarem o mercado.

```python
### 6. Top 10 Bandeiras que mais venderam 2019 a 2023

# Agrupando e somando os dados
bandeiras_valor = df.groupby('bandeira')['valor_venda'].sum().sort_values(ascending=False).head(10)

# Criando o gráfico
plt.figure(figsize=(10,6))
ax = sns.barplot(x=bandeiras_valor.values, y=bandeiras_valor.index, palette='viridis')
plt.title('Top 10 Bandeiras que mais venderam 2019 a 2023')
plt.xlabel('Valor Vendido (R$)')
plt.ylabel('Bandeira')

# Adicionando rótulos de dados
for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

plt.show()
```

As bandeiras mais lucrativas são destacadas, mostrando a força das principais marcas no mercado de combustíveis.
